```python
import time
def hi():
    for i in range(100):
        print(i,"hi")
        print(1/0)
        time.sleep(1)
def hello():
    for i in range(100):
        print(i,"hello")
        time.sleep(1)
```

```python
from threading import Thread
```

```python
t1=Thread(target=hi,name="krishna")
t2=Thread(target=hello,name="pintu")
```

```python
t1.start()
t2.start()
```
